```python
import pandas as pd
from google.colab import files
import seaborn as sns
import matplotlib.pyplot as plt

path='/content/FRvideos_cc50_202101.csv'

Francia_data = pd.read_csv(path)
```

```python
# Obtener los nombres de las columnas
list(Francia_data.columns)
```

```python
# Ver la estructura de los datos
print(Francia_data.info())
```

```python
# Mostrar los tipos de datos de cada columna
print(Francia_data.dtypes)
```

```python
# Mostrar los primeros registros
Francia_data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(Francia_data.index[:10], Francia_data['views'].head(10), label='Vistas')
plt.bar(Francia_data.index[:10], Francia_data['likes'].head(10), label='Likes')
plt.xlabel('Videos')
plt.ylabel('Cantidad')
plt.title('Comparación de Vistas y Likes de los Primeros 10 Videos')
plt.legend()
plt.show()
```

### Verificar la calidad de los datos



*   Identificar valores nulos o faltantes





```python
valores_nulos = Francia_data.isnull().sum()
porcentaje_valores_nulos = (Francia_data.isnull().sum() / len(Francia_data)) * 100
valores_nulos_info = pd.concat([valores_nulos, porcentaje_valores_nulos], axis=1)
valores_nulos_info.columns = ['Valores Nulos', '% de Valores Nulos']
print(valores_nulos_info)
```



*   Analizar los valores atipicos



```python
# Analizar valores atípicos para columnas numéricas
# Crear boxplots para 'lat' y 'lon' para identificar valores atípicos
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=Francia_data['lat'])
plt.title('Boxplot para Latitud (lat)')

plt.subplot(1, 2, 2)
sns.boxplot(y=Francia_data['lon'])
plt.title('Boxplot para Longitud (lon)')

plt.tight_layout()
plt.show()
```
